```python
import PyPDF2
import re
```

```python
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text
```

```python
def remove_sections(text, sections_to_remove):
    for section in sections_to_remove:
        pattern = re.compile(section, re.DOTALL)
        text = re.sub(pattern, '', text)
    return text
```

```python
def extract_financial_data(text):
    # Example: Extracting financial data related to revenue, profit, etc.
    revenue_pattern = re.compile(r"Revenue[^:]*:?\s*(\d[\d,\.]*)")
    profit_pattern = re.compile(r"Profit[^:]*:?\s*(\d[\d,\.]*)")
    
    revenue = revenue_pattern.findall(text)
    profit = profit_pattern.findall(text)
    
    financial_data = {
        "Revenue": revenue,
        "Profit": profit
    }
    return financial_data
```

```python
def main():
    pdf_path = 'SanstarLimited_RHP.pdf'  # Replace with your path
    pdf_text = extract_text_from_pdf(pdf_path)
    print(pdf_text)
    
    # Define the sections to remove using start and end keywords or phrases
    sections_to_remove = [
        r"SECTION I – GENERAL.*?SECTION II - ",  # Example: remove from "SECTION I – GENERAL" to "SECTION II"
        r"RISK FACTORS.*?INTRODUCTION"  # Example: remove from "RISK FACTORS" to "INTRODUCTION"
    ]
    
    cleaned_text = remove_sections(pdf_text, sections_to_remove)
    
    financial_data = extract_financial_data(cleaned_text)
    
    print("Extracted Financial Data:")
    for key, value in financial_data.items():
        print(f"{key}: {value}")

if __name__ == "__main__":
    main()
```
